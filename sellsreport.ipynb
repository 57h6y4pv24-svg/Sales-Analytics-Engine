```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Fecha': [
        '2024-01-15','2024-01-16','2024-02-10','2024-02-15',
        '2024-03-05','2024-03-20','2024-04-10','2024-04-15',
        '2024-05-01','2024-05-20'
    ],
    'Cliente': [
        'Juan','María','Pedro','María',
        'Juan','Ana','Carlos','Ana',
        'Pedro','María'
    ],
    'Producto': [
        'Shampoo','Protector Solar','Jabón Facial','Shampoo',
        'Protector Solar','Jabón Facial','Shampoo','Protector Solar',
        'Jabón Facial','Shampoo'
    ],
    'Categoría': [
        'Cuidado Personal','Skin Care','Skin Care','Cuidado Personal',
        'Skin Care','Skin Care','Cuidado Personal','Skin Care',
        'Skin Care','Cuidado Personal'
    ],
    'Cantidad': [2,1,3,2,1,2,4,2,1,3],
    'Precio Unitario': [45,120,60,45,120,60,45,120,60,45]
}

df = pd.DataFrame(data)
df['Total Venta'] = df['Cantidad'] * df['Precio Unitario']

productos = df.groupby('Producto')['Cantidad'].sum().sort_values(ascending=False)
clientes = df.groupby('Cliente')['Total Venta'].sum().sort_values(ascending=False)
categorias = df.groupby('Categoría')['Total Venta'].sum().sort_values(ascending=False)

df['Fecha'] = pd.to_datetime(df['Fecha'])
df['Mes'] = df['Fecha'].dt.month_name()
ventas_mes = df.groupby('Mes')['Total Venta'].sum()

print("Producto más vendido:", productos.idxmax())
print("Cliente que más compró:", clientes.idxmax())
print("Categoría con más ingresos:", categorias.idxmax())
print("Ventas totales:", df['Total Venta'].sum())
```

```python
with pd.ExcelWriter('reporte_ventas.xlsx') as writer:
    df.to_excel(writer, sheet_name='Datos Originales', index=False)
    productos.to_excel(writer, sheet_name='Productos')
    clientes.to_excel(writer, sheet_name='Clientes')
    categorias.to_excel(writer, sheet_name='Categorias')
    ventas_mes.to_excel(writer, sheet_name='Ventas por Mes')

print("Reporte creado correctamente")
```
